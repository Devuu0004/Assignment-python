1.What is the difference between a function and a method in Python?
  - In Python, the difference between a function and a method lies mainly in how they are defined and how they are called.
    -  Function:
       A function is a block of code that is defined using the def keyword and can be called independently.

       It may or may not take parameters and may or may not return a value.

       Functions are usually not associated with any specific object.
    - Method:
      A method is essentially a function, but it is associated with an object (an instance of a class) or a class itself.

      Methods are defined inside a class and are called on objects or classes.

      A method has at least one special parameter (usually self for instance methods), which refers to the object on which it is being called.

2.Explain the concept of function arguments and parameters in Python.
  - In Python, function arguments and parameters refer to the way values are passed to a function when it is called and how those values are received by the function.

   - Function Parameters:
     Parameters are variables that are defined in the function signature (the function definition).

5.What are iterators in Python and how do they differ from iterables?
  - In Python, iterators and iterables are closely related concepts that deal with looping or traversing through a collection (like a list, tuple, or dictionary). However, they serve different purposes in the context of iteration.

  -  Iterable:
     An iterable is any Python object capable of returning its members one at a time, permitting it to be iterated over in a for-loop. An iterable is an object that implements the __iter__() method or the __getitem__() method.

     Characteristics of an Iterable:
     It is an object that can be iterated (looped) over.

     It contains a sequence of items (e.g., a list, string, dictionary, etc.).

     It must implement the __iter__() method, which returns an iterator, or alternatively, it can implement the __getitem__() method, which allows indexing.

  - Iterator

     An iterator is an object that represents a stream of data; it does the actual work of iterating over the elements of an iterable. An iterator must implement two special methods:

     __iter__(): Returns the iterator object itself (this is required for the object to be an iterator).

     __next__(): Returns the next item in the sequence, and raises a StopIteration exception when no more items are available.

     Characteristics of an Iterator:
     It is an object that keeps track of the current state of iteration.

     It is used to access elements one at a time from an iterable.

     It has a __next__() method, which retrieves the next value and raises StopIteration when the iteration is finished.

6.Explain the concept of generators in Python and how they are defined.
  - In Python, a generator is a special type of iterator that allows you to iterate through a sequence of values one at a time, but without storing the entire sequence in memory. Generators are defined using a function that contains one or more yield expressions, and they produce values lazily, meaning the next value is generated only when requested.

  Generators are a more memory-efficient alternative to using lists or other data structures when you need to handle large datasets or sequences.

  Key Characteristics of Generators:
  Lazy Evaluation: A generator yields values one by one when requested, and it doesn't compute all the values upfront. This allows you to handle potentially large or infinite sequences of data without consuming too much memory.

  State Preservation: Unlike regular functions that start fresh every time they're called, a generator function "remembers" its state between calls. This allows it to pick up where it left off and continue generating values when needed.

  yield Keyword: Instead of return, a generator uses the yield keyword to produce values one at a time. When the generator's next() function is called, it resumes execution from where it last yielded a value.

  StopIteration: When all the values have been generated, a StopIteration exception is automatically raised, signaling that there are no more values to yield.

  - How to Define a Generator:
     Generators are defined like normal functions, but instead of return, they use the yield statement to return values one at a time.

9.Explain the purpose and usage of the `map()` function in Python.

    function to all items in an iterable (like a list, tuple, or any other iterable) and return a new iterable (usually a map object) with the results.

  - Purpose: The map() function applies a function to all items in an iterable
    (or iterables) and returns a map object containing the results.

    Use Cases: It's often used when you need to apply a transformation to all elements of a sequence without writing an explicit loop.

    Advantages: More concise code, and it can be more efficient in terms of memory usage due to its lazy evaluation.

    When to Use: Use map() when you want to apply a function to every item of an iterable, especially in simple cases or when combined with lambda functions.

10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

 - The map(), reduce(), and filter() functions in Python are all used to process iterables, but they each serve a different purpose and work in unique ways. Here's a breakdown of the differences:

  1. map() Function
     Purpose: The map() function applies a given function to each item of an iterable (or multiple iterables) and returns a new iterable (a map object) with the results.

     Output: It returns an iterable (map object), which can be converted to a list, tuple, etc., containing the results of applying the function to each item in the input iterable(s).

     Use Case: When you want to transform each item in an iterable using a function.

  2. reduce() Function
     Purpose: The reduce() function (from the functools module) applies a given function cumulatively to the items of an iterable. It reduces the iterable to a single value by applying the function to pairs of items successively.

     Output: It returns a single value that results from applying the function cumulatively to the iterable's items.

     Use Case: When you want to reduce a sequence of values to a single value (e.g., summing a list, multiplying values together, etc.).

 3. filter() Function
     Purpose: The filter() function filters elements of an iterable based on a condition defined by a function. It returns a new iterable with only the elements that satisfy the condition (i.e., for which the function returns True).

     Output: It returns an iterable (filter object) that contains only the elements for which the function returns True. The result is typically converted to a list, tuple, etc.

     Use Case: When you want to filter elements in an iterable based on some condition.

```python
#1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
  #the list.

def sum_of_even_numbers(numbers):
    # Use filter to get only even numbers, then sum them up
    even_numbers = filter(lambda x: x % 2 == 0, numbers)
    return sum(even_numbers)

# Example usage:
numbers = [47, 11, 42, 13, 8, 16]
result = sum_of_even_numbers(numbers)
print("Sum of even numbers:", result)
```

```python
#4.Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(number):
    if number < 2 or number > 200:
        return False  # Prime numbers are greater than or equal to 2 and less than or equal to 200.

    for i in range(2, int(number ** 0.5) + 1):
        if number % i == 0:
            return False  # If the number is divisible by any number other than 1 and itself, it's not prime.

    return True  # If no divisors are found, the number is prime.

# Example usage:
numbers = range(1, 201)
prime_numbers = [num for num in numbers if is_prime(num)]
print("Prime numbers from 1 to 200:", prime_numbers)
```

```python
#5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
#terms.
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms  # Number of terms to generate
        self.current = 0  # The current term number
        self.a, self.b = 0, 1  # The first two numbers in the Fibonacci sequence

    def __iter__(self):
        return self  # Returns the iterator itself

    def __next__(self):
        if self.current >= self.terms:
            raise StopIteration  # Stop the iteration once the specified terms are generated
        else:
            result = self.a  # Store the current Fibonacci number
            self.a, self.b = self.b, self.a + self.b  # Update to the next two Fibonacci numbers
            self.current += 1  # Increment the term counter
            return result  # Return the current Fibonacci number

# Example usage:
fibonacci = FibonacciIterator(10)  # Generate the first 10 Fibonacci numbers
for num in fibonacci:
    print(num)
```

```python
#6.Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_2(5):  # Generate powers of 2 up to 2^5
    print(power)
```

```python
#8.Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# Sample list of tuples
tuples_list = [(1, 3), (4, 1), (2, 2), (5, 5)]

# Sort the list of tuples based on the second element of each tuple using a lambda function
sorted_tuples = sorted(tuples_list, key=lambda x: x[1])

# Output the sorted list
print(sorted_tuples)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 30, 40, 100]

# Use map() to apply the conversion function to each element in the list
fahrenheit_temperatures = map(celsius_to_fahrenheit, celsius_temperatures)

# Convert the result to a list and print the temperatures in Fahrenheit
print(list(fahrenheit_temperatures))
```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

# Function to check if a character is not a vowel
def is_not_vowel(char):
    vowels = 'aeiouAEIOU'
    return char not in vowels

# Input string
input_string = "Hello, World!"

# Use filter() to remove vowels from the string
filtered_string = filter(is_not_vowel, input_string)

# Convert the filtered result to a string and print it
result = ''.join(filtered_string)
print("String without vowels:", result)
```
